Q1.What is hypothesis testing in statistics?

.Hypothesis Testing is a type of statistical analysis in which you put your assumptions about a population parameter to the test. It is used to estimate the relationship between 2 statistical variables. When performing hypothesis testing, you must understand different data types, such as nominal data.

Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

.In hypothesis testing, the null hypothesis (H0) proposes no effect or relationship, while the alternative hypothesis (H1 or Ha) suggests an effect or relationship exists, and is what the researcher aims to prove.

Here's a more detailed breakdown:

Null Hypothesis (H0):
This is the default assumption, stating there's no difference or effect between groups or variables.
It's often the hypothesis a researcher tries to disprove or reject.

Examples:

"There is no difference in the average test scores between two groups."
"There is no relationship between smoking and lung cancer.

Alternative Hypothesis (H1 or Ha):

This is the statement that contradicts the null hypothesis, suggesting there is a difference or effect.
It's the hypothesis the researcher is trying to prove or support.

Examples:
"There is a difference in the average test scores between two groups."
"There is a relationship between smoking and lung cancer."


Q3.What is the significance level in hypothesis testing, and why is it important?.

.In hypothesis testing, the significance level (often denoted as α) is the probability of rejecting a true null hypothesis, also known as a Type I error. It's important because it sets the threshold for determining statistical significance and helps researchers control the risk of making false conclusions.

Q4.What does a P-value represent in hypothesis testing?

.In hypothesis testing, a p-value represents the probability of observing results as extreme as, or more extreme than, the ones obtained from a sample, assuming the null hypothesis is true.

Q5.How do you interpret the P-value in hypothesis testing?

.Small p-value (e.g., < 0.05): This indicates that the observed results are unlikely to have occurred by chance if the null hypothesis were true. This provides evidence to reject the null hypothesis in favor of the alternative hypothesis.

Large p-value (e.g., > 0.05): This suggests that the observed results are likely to have occurred by chance if the null hypothesis is true. In this case, you would fail to reject the null hypothesis.

Q6.What are Type 1 and Type 2 errors in hypothesis testing?

.In hypothesis testing, a Type I error (false positive) occurs when you reject a true null hypothesis, while a Type II error (false negative) occurs when you fail to reject a false null hypothesis.

Q7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

.Two-tailed test:

Also known as a non-directional test.

Tests for a difference, regardless of direction (either an increase or decrease).

The alternative hypothesis doesn't specify a direction.

Example: Testing if a new drug has an effect on blood pressure (without specifying whether it will increase or decrease it).

One-tailed test:

Also known as a directional test.

Tests for a difference in a specific direction (either an increase or a decrease).

The alternative hypothesis specifies a direction (e.g., greater than or less than).


Example: Testing if a new fertilizer increases crop yield (but not if it decreases it).

Q8.What is the Z-test, and when is it used in hypothesis testing?

Purpose:

In hypothesis testing, a z-score indicates how many standard deviations a data point is from the mean, helping determine if the observed data is likely to come from the hypothesized population.

Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import math
from scipy import stats


def z_test(sample_mean, population_mean, population_std, sample_size):

    z_score = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value


def interpret_results(z_score, p_value, alpha=0.05):
    print(f"Z-score: {z_score}")
    print(f"P-value: {p_value}")

    if p_value < alpha:
        print(f"Since the p-value ({p_value}) is less than the significance level ({alpha}), we reject the null hypothesis.")
        print("This suggests that there is a significant difference between the sample mean and the population mean.")
    else:
        print(f"Since the p-value ({p_value}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
        print("This suggests that there is no significant difference between the sample mean and the population mean.")


sample_mean = 105
population_mean = 100
population_std = 15
sample_size = 30


z_score, p_value = z_test(sample_mean, population_mean, population_std, sample_size)


interpret_results(z_score, p_value)
```

Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math


def z_test(sample_data, population_mean, population_std, alpha=0.05):

    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)


    z_score = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))


    z_critical = stats.norm.ppf(1 - alpha / 2)
    print(f"Critical Z-scores: ±{z_critical}")

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))


    print(f"Sample Mean: {sample_mean}")
    print(f"Z-Score: {z_score}")
    print(f"P-value: {p_value}")


    if abs(z_score) > z_critical:
        print(f"Since the Z-score ({z_score}) is outside the critical region (±{z_critical}), we reject the null hypothesis.")
        print("This suggests that there is a significant difference between the sample mean and the population mean.")
    else:
        print(f"Since the Z-score ({z_score}) is inside the critical region (±{z_critical}), we fail to reject the null hypothesis.")
        print("This suggests that there is no significant difference between the sample mean and the population mean.")


    plot_decision_region(z_score, z_critical)


def plot_decision_region(z_score, z_critical):

    z_values = np.linspace(-4, 4, 1000)
    pdf_values = stats.norm.pdf(z_values)

    plt.figure(figsize=(10, 6))
    plt.plot(z_values, pdf_values, label="Standard Normal Distribution", color='blue')


    plt.fill_between(z_values, pdf_values, where=(z_values <= -z_critical) | (z_values >= z_critical), color='red', alpha=0.3, label="Rejection Regions (α = 0.05)")


    plt.axvline(x=z_critical, color='red', linestyle='dashed', label=f"Critical Z = ±{z_critical}")
    plt.axvline(x=-z_critical, color='red', linestyle='dashed')

    plt.axvline(x=z_score, color='green', linestyle='dashed', label=f"Observed Z = {z_score:.2f}")

    plt.title("Two-Tailed Z-Test Decision Region")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend()


    plt.grid(True)
    plt.show()


np.random.seed(42)
sample_size = 50
population_mean = 100
population_std = 15


sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

z_test(sample_data, population_mean, population_std)
```

Q5.
 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(population_mean, population_std, alternative_mean, sample_size, alpha=0.05):

    z_critical = norm.ppf(1 - alpha / 2)


    sample_std = population_std / np.sqrt(sample_size)

    rejection_region_left = population_mean - z_critical * sample_std
    rejection_region_right = population_mean + z_critical * sample_std

    x = np.linspace(population_mean - 4 * sample_std, population_mean + 4 * sample_std, 1000)


    y_null = norm.pdf(x, population_mean, sample_std)

    y_alt = norm.pdf(x, alternative_mean, sample_std)


    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label=f"Null Hypothesis Distribution (H0)", color="blue")
    plt.plot(x, y_alt, label=f"Alternative Hypothesis Distribution (H1)", color="green")


    plt.fill_between(x, 0, y_null, where=(x <= rejection_region_left) | (x >= rejection_region_right), color="red", alpha=0.3, label="Type 1 Error (α)")


    plt.fill_between(x, 0, y_alt, where=(x >= rejection_region_left) & (x <= rejection_region_right), color="yellow", alpha=0.3, label="Type 2 Error (β)")


    plt.axvline(rejection_region_left, color="red", linestyle="dashed", label=f"Critical Value (-Z_c) = {rejection_region_left:.2f}")
    plt.axvline(rejection_region_right, color="red", linestyle="dashed", label=f"Critical Value (+Z_c) = {rejection_region_right:.2f}")

    plt.title(f"Type 1 and Type 2 Errors in Hypothesis Testing\nα = {alpha}, Sample Size = {sample_size}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()

    plt.grid(True)
    plt.show()


population_mean = 100
population_std = 15
alternative_mean = 105
sample_size = 30
alpha = 0.05

visualize_errors(population_mean, population_std, alternative_mean, sample_size, alpha)
```

Q6. Write a Python program to perform an independent T-test and interpret the results

```python
import numpy as np
from scipy import stats


def independent_t_test(group1, group2, alpha=0.05):

    mean1 = np.mean(group1)
    mean2 = np.mean(group2)
    std1 = np.std(group1, ddof=1)
    std2 = np.std(group2, ddof=1)
    n1 = len(group1)
    n2 = len(group2)


    t_statistic = (mean1 - mean2) / np.sqrt((std1**2 / n1) + (std2**2 / n2))


    df = ((std1**2 / n1 + std2**2 / n2) ** 2) / \
         (((std1**2 / n1) ** 2) / (n1 - 1) + ((std2**2 / n2) ** 2) / (n2 - 1))


    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))


    print(f"Group 1 Mean: {mean1}, Group 2 Mean: {mean2}")
    print(f"Group 1 Standard Deviation: {std1}, Group 2 Standard Deviation: {std2}")
    print(f"T-statistic: {t_statistic}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value}")


    if p_value < alpha:
        print(f"Since the p-value ({p_value}) is less than the significance level ({alpha}), we reject the null hypothesis.")
        print("This suggests that there is a significant difference between the two groups.")
    else:
        print(f"Since the p-value ({p_value}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
        print("This suggests that there is no significant difference between the two groups.")


np.random.seed(42)
group1 = np.random.normal(loc=100, scale=15, size=30)
group2 = np.random.normal(loc=110, scale=15, size=30)


independent_t_test(group1, group2, alpha=0.05)
```

Q9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats


def confidence_interval(sample, confidence_level=0.95):

    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    sample_size = len(sample)


    df = sample_size - 1


    t_critical = stats.t.ppf((1 + confidence_level) / 2, df)


    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))


    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error


    return lower_bound, upper_bound


np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)


lower_bound, upper_bound = confidence_interval(sample, confidence_level=0.95)


print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

Q11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python

def bayesian_inference(prior, likelihood, false_positive_rate, marginal_likelihood):
    """
    Compute the posterior probability using Bayes' Theorem.

    :param prior: Prior probability of hypothesis (probability of having the disease).
    :param likelihood: Probability of observing the evidence given the hypothesis (True Positive Rate).
    :param false_positive_rate: Probability of observing the evidence when the hypothesis is false (False Positive Rate).
    :param marginal_likelihood: Probability of the evidence (marginal likelihood).
    :return: Posterior probability
    """
    posterior = (likelihood * prior) / marginal_likelihood
    return posterior


prior_disease = 0.01
likelihood_positive_given_disease = 0.95
false_positive_rate = 0.05
marginal_likelihood = (likelihood_positive_given_disease * prior_disease) + (false_positive_rate * (1 - prior_disease))


posterior_disease_given_positive_test = bayesian_inference(prior_disease, likelihood_positive_given_disease, false_positive_rate, marginal_likelihood)


print(f"The probability of having the disease given a positive test result is: {posterior_disease_given_positive_test:.4f}")
```

```python
import numpy as np


def calculate_expected_frequencies(observed_data):

    row_totals = np.sum(observed_data, axis=1)
    column_totals = np.sum(observed_data, axis=0)
    grand_total = np.sum(observed_data)

    expected_frequencies = np.outer(row_totals, column_totals) / grand_total

    return expected_frequencies


observed_data = np.array([[30, 10],
                          [20, 40]])


expected_data = calculate_expected_frequencies(observed_data)


print("Observed Data:")
print(observed_data)

print("\nExpected Frequencies:")
print(expected_data)
```

Q16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=5, size=50)
sample2 = np.random.normal(loc=10, scale=7, size=40)


var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

F_statistic = var1 / var2


df1 = len(sample1) - 1
df2 = len(sample2) - 1


p_value = 1 - f.cdf(F_statistic, df1, df2)


print(f"Variance of Sample 1: {var1}")
print(f"Variance of Sample 2: {var2}")
print(f"F-statistic: {F_statistic}")
print(f"p-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject H0: The variances of the two samples are significantly different.")
else:
    print("\nFail to reject H0: The variances of the two samples are not significantly different.")
```

Q19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def check_anova_assumptions(groups):

    print("Checking Normality of Groups:")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1}: Shapiro-Wilk Test p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i+1}: The data is NOT normally distributed (p < 0.05)")
        else:
            print(f"Group {i+1}: The data is normally distributed (p >= 0.05)")
    print()


    print("Checking for Independence: Assumed based on study design.")
    print("No specific statistical test is conducted here, but scatter plots can be used for time series data.")
    print()


    print("Checking for Equal Variance (Homoscedasticity) using Levene's Test:")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's Test p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("The variances are significantly different (p < 0.05), assumption of equal variance is violated.")
    else:
        print("The variances are not significantly different (p >= 0.05), assumption of equal variance is met.")
    print()


    plt.figure(figsize=(10, 6))
    for i, group in enumerate(groups):
        plt.subplot(1, len(groups), i+1)
        plt.hist(group, bins=15, alpha=0.7, color='blue', edgecolor='black')
        plt.title(f'Group {i+1} Histogram')
        plt.xlabel('Values')
        plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()


np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=12, size=30)
group3 = np.random.normal(loc=60, scale=15, size=30)

check_anova_assumptions([group1, group2, group3])
```

Q20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(42)


treatment = np.array(['A', 'B', 'C'])
time = np.array(['T1', 'T2'])


data = []
for t in treatment:
    for tm in time:

        values = np.random.normal(loc=50 + 5*(treatment.tolist().index(t) + 1) + 3*(time.tolist().index(tm) + 1), scale=10, size=20)
        data.extend(list(zip([t]*20, [tm]*20, values)))


df = pd.DataFrame(data, columns=['Treatment', 'Time', 'Value'])

print(df.head())

model = ols('Value ~ C(Treatment) + C(Time) + C(Treatment):C(Time)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\nTwo-way ANOVA Results:")
print(anova_table)

alpha = 0.05
if anova_table['PR(>F)']['C(Treatment):C(Time)'] < alpha:
    print("\nThere is a significant interaction effect between Treatment and Time.")
else:
    print("\nThere is no significant interaction effect between Treatment and Time.")


plt.figure(figsize=(8, 6))
sns.pointplot(x='Time', y='Value', hue='Treatment', data=df, markers=["o", "s", "D"], linestyles=["-", "--", "-."], dodge=True)
plt.title('Interaction Plot for Treatment and Time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Treatment', y='Value', hue='Time', data=df)
plt.title('Boxplot of Treatment and Time Effects')
plt.xlabel('Treatment')
plt.ylabel('Value')
plt.show()
```

Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)
population_variance = 25


sample_data = np.random.normal(loc=0, scale=np.sqrt(population_variance), size=sample_size)



sample_variance = np.var(sample_data, ddof=1)
sample_size = len(sample_data)


chi_square_statistic = (sample_size - 1) * sample_variance / population_variance

df = sample_size - 1
alpha = 0.05


p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))


print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-Square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from 25.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference from the population variance of 25.")
```

Q25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats


x1 = 80
n1 = 200


x2 = 60
n2 = 180

p1 = x1 / n1
p2 = x2 / n2

P = (x1 + x2) / (n1 + n2)

se = np.sqrt(P * (1 - P) * (1 / n1 + 1 / n2))

Z = (p1 - p2) / se

p_value = 2 * (1 - stats.norm.cdf(abs(Z)))


print(f"Sample Proportion for Group 1: {p1:.4f}")
print(f"Sample Proportion for Group 2: {p2:.4f}")
print(f"Z-statistic: {Z:.4f}")
print(f"P-value: {p_value:.4f}")


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions in the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the proportions of the two groups.")
```

Q27.Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)

observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=100)

observed_frequencies = np.array([np.sum(observed_data == i) for i in range(1, 7)])


expected_frequencies = np.ones(6) * (100 / 6)


chi_square_statistic = np.sum((observed_frequencies - expected_frequencies) ** 2 / expected_frequencies)


df = len(observed_frequencies) - 1

p_value = 1 - stats.chi2.cdf(chi_square_statistic, df)


print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies: {expected_frequencies}")
print(f"Chi-Square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")


categories = [1, 2, 3, 4, 5, 6]

plt.bar(categories, observed_frequencies, label='Observed', alpha=0.6, color='blue', width=0.4, align='center')
plt.bar(categories, expected_frequencies, label='Expected', alpha=0.6, color='red', width=0.4, align='edge')

plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Chi-Square Goodness of Fit Test')
plt.legend()
plt.show()
```
